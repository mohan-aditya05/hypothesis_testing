# Alternate hypothesis: Promotion is effective.

```python
from statsmodels.stats.weightstats import ztest
z_statistic, p_value = ztest(x1=data1[data1['insert'] == 'New Promotion']['dollars'], value=data1[data1['insert'] == 'Standard']['dollars'].mean())
print('Z-statistic is :{}'.format(z_statistic))
print('P-value is :{}'.format(p_value))
```

```python
data4=pd.read_csv('..//data/sample_survey.csv')
data4
```

```python
data4.columns
```

```python
import scipy.stats as sp
def compute_freq_chi2(x,y):
    """This function will compute frequency table of x an y
    Pandas Series, and use the table to feed for the contigency table
    
    Parameters:
    -------
    x,y : Pandas Series, must be same shape for frequency table
    
    Return:
    -------
    None. But prints out frequency table, chi2 test statistic, and 
    p-value
    """
    freqtab = pd.crosstab(x,y)
    print("Frequency table")
    print("============================")
    print(freqtab)
    print("============================")
    chi2,pval,dof,expected = sp.chi2_contingency(freqtab)
    print("ChiSquare test statistic: ",chi2)
    print("p-value: ",pval)
    return
```

# Null Hypothesis: *There is no relationship between labour force status and marital status.*
# Alternate Hypothesis: *A relationship exists between labour force status and marital status.*

***Here we apply Chi-square test of independence!***

```python
compute_freq_chi2(data4.marital, data4.wrkstat)
```

* Since p-value is less than 0.05 for 95% confidence interval. 
* Therefore, we reject the null hypothesis. 

# Null Hypothesis: *There is no relationship between educational qualification and marital status.*
# Alternate Hypothesis: *A relationship exists between educational qualification and marital status.*

```python
compute_freq_chi2(data4.degree, data4.marital)
```

* Since p-value is less than 0.05 for 95% confidence interval. 
* Therefore, we reject the null hypothesis. 

# Null Hypothesis: *Happiness is not driven by both income and marital status.*
# Alternate Hypothesis: *Happiness is driven by both income and /or marital status.*

```python
compute_freq_chi2(data4.happy, data4.marital)
```

```python
compute_freq_chi2(data4.happy, data4.income)
```

* Since p-value is less than 0.05 for both chi-square independence tests for 95% confidence interval. 
* Therefore, we reject the null hypothesis. 
* Therefore, happiness depends on both marital status and income.
